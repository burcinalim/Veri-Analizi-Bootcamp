## Outlier Analysis

Data values that are usually statistically extreme are called outliers. We can leave data as is, delete the outliers, review them,and  transform / limit them. We should separate numerical variables from categorical to find outliers because categorical variables don't have statistical significance.

```python
# Select numerical columns
numeric_df = my_data.select_dtypes(include=['number'])

# Drop unnecessary columns. (We don't need to analyze id column.)
numeric_df = numeric_df.drop(['id'], axis=1)
```

```python
plt.figure(figsize=(16, 8))
sns.boxplot(data=numeric_df)
plt.title("Outliers")
plt.xticks(rotation=60)
plt.show()
```

The above boxplot shows what categories have outliers. Especially **Flight Distance**, **Departure Delay in Minutes**,and **Arrival Delay in Minutes** categories are pointed out. These dimensions may be meaningful for flight reliability. They may not need to be deleted, even if they appear to be contradictory.

```python
# There is a small variance in Checkin Service dimension in outlier analysis. So, this dimension can be looked at in detail.(According to the boxplot, there is "1 point" which is in the score range, it can be kept within the data. )

sns.boxplot(x=numeric_df['Checkin service'])
plt.title("Checkin Service Outlier Analysis")
plt.show()
```

### Analysis of Numerical Variables

```python
# Select just survey's satisfaction dimensions.

survey_dimensions = ['Inflight wifi service', 'Departure/Arrival time convenient', 'Ease of Online booking','Gate location','Food and drink','Online boarding','Seat comfort','Inflight entertainment','On-board service','Leg room service','Baggage handling','Checkin service','Inflight service','Cleanliness']
```

#### Overall Average Satisfaction Score

```python
# Create a new column to show avg. satisfaction scores of each customers' evaluation.
my_data['Average_Satisfaction_Score'] = my_data[survey_dimensions].mean(axis=1)

# Calculate overall average score.
overall_avg = my_data['Average_Satisfaction_Score'].mean()
print("Overall Satisfaction Score:", round(overall_avg, 2))
```

```python
# Visualize average satisfaction score with hisplot.
sns.histplot(my_data['Average_Satisfaction_Score'], bins=20, kde=True)
plt.axvline(overall_avg, color='red', linestyle='--', linewidth=2, label=f'Overall Satisfaction Score: {overall_avg:.2f}')
plt.legend()
plt.title("Distribution of Average Satisfaction Score")
plt.xlabel("Average Score")
plt.ylabel("Customer Count")
plt.show()
```

#### Average Satisfaction Scores Chart (Based on Survey Dimensions) 

```python
# Mean scores for each dimension of the survey.

avg_scores_of_dimensions = my_data[survey_dimensions].mean(axis=0)
```

```python
mean_scores = avg_scores_of_dimensions.sort_values(ascending=True)
mean_scores.plot(kind='barh', figsize=(10,6), color='skyblue')
plt.title("Average Satisfaction Scores of Dimensions")
plt.xlabel("Average Score")

# to write values of dimensions into the chart
for i, (index, value) in enumerate(mean_scores.items()):
    plt.text(value + 0.05, i, f"{value:.2f}", va='center', fontsize=9)

plt.tight_layout()
plt.show()
```

According to the chart above, 9 out of 14 dimensions are above the overall score(3.24).
