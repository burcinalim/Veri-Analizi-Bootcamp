```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data_1 = pd.read_csv("accidents_2005_to_2007.csv", low_memory=False)
data_2 = pd.read_csv("accidents_2009_to_2011.csv", low_memory=False)
data_3 = pd.read_csv("accidents_2012_to_2014.csv", low_memory=False)
meta_data = pd.concat([data_1,data_2,data_3], ignore_index=True)
meta_data.head()
```

```python
meta_data.describe().T
```

```python
meta_data.describe(include = 'object')
```

## MISSING VALUES

```python
meta_data.isnull().sum()
```

```python
missing_ratio = meta_data.isnull().mean().sort_values(ascending=False)
missing_ratio = missing_ratio[missing_ratio > 0] *100
print(missing_ratio)
```

**Columns with very high missingness rate (>95%)**

```python
# Filter columns missing more than 95
cols_to_drop = missing_ratio[missing_ratio > 95].index
cols_to_drop
```

```python
# Delete these columns from the data set
meta_data = meta_data.drop(columns = cols_to_drop)
```

**Low level of deficiency (<1%)**

```python
cols_to_drop_2 = missing_ratio[missing_ratio < 1].index
cols_to_drop_2
```

```python
meta_data = meta_data.dropna(subset = cols_to_drop_2)
```

**Moderate deficiency (10%-50%)**

These can be important categorical variables. If the missingness is not random, it carries information. We should add missing indicators before do any process.

```python
# 1. Adding Missing Indicator

columns_with_flags = ['Junction_Control', 'LSOA_of_Accident_Location']

for col in columns_with_flags:
    meta_data[f'Missing_{col}'] = meta_data[col].isnull().astype(int)

# 2. Filling the missing values

#### Fill with "None" 
cols_fill_none = ['Junction_Control', 'LSOA_of_Accident_Location']

for col in cols_fill_none:
    meta_data[col] = meta_data[col].fillna('None')
```

```python
# To check the latest condition
meta_data.isnull().sum()
```

## OUTLIER ANALYSIS

```python
outlier_counts = {}
num_cols = meta_data.select_dtypes(include=['float','int']).columns

for col in num_cols:
    Q1 = meta_data[col].quantile(0.25)
    Q3 = meta_data[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = meta_data[(meta_data[col] < lower_bound) | (meta_data[col] > upper_bound)]
    
    if not outliers.empty:
        outlier_counts[col] = outliers.shape[0]

# SonuÃ§larÄ± sÄ±ralÄ± ÅŸekilde gÃ¶ster
outlier_df = pd.DataFrame.from_dict(outlier_counts, orient='index', columns=['Outlier Count'])
outlier_df = outlier_df.sort_values(by='Outlier Count', ascending=False)

print("Columns with outliers and outlier counts:")
print(outlier_df)
```

```python
num_cols = len(outlier_counts)
cols_per_row = 3  # Her satÄ±ra 3 grafik
rows = (num_cols // cols_per_row) + 1

fig, axes = plt.subplots(rows, cols_per_row, figsize=(cols_per_row*5, rows*3))

axes = axes.flatten()  # DÃ¼zleÅŸtir ki kolay eriÅŸelim

for i, col in enumerate(outlier_counts.keys()):
    sns.boxplot(data=meta_data, x=col, ax=axes[i])
    axes[i].set_title(f"{col}")

# KullanÄ±lmayan subplot'larÄ± kapat
for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
```

## VISUALIZATION

**How has changing traffic flow impacted accidents?** <br>

*We can answer this question by comparing total accidents and total vehicle count on the road.*

```python
traffic_flow = pd.read_csv("ukTrafficAADF.csv", low_memory=False)
traffic_flow.head(3)
```

```python

vehicles_per_year = traffic_flow.groupby("AADFYear")["AllMotorVehicles"].sum().reset_index(name="Total_Vehicles")
accidents_per_year = meta_data.groupby("Year").size().reset_index(name="Total_Accidents")

# 2. To merge common years : (inner join)
merged_df = pd.merge(vehicles_per_year, accidents_per_year, left_on="AADFYear", right_on="Year", how="inner")

# Graph
fig, ax1 = plt.subplots(figsize=(12, 6))

# Left axis: Number of vehicles
ax1.plot(merged_df["Year"], merged_df["Total_Vehicles"], color='blue', marker='o', label='Total Vehicles')
ax1.set_xlabel("Year")
ax1.set_ylabel("Total Vehicles", color='blue')
ax1.tick_params(axis='y', labelcolor='blue')
ax1.grid(True)

# Right axis: Number of accidents
ax2 = ax1.twinx()
ax2.plot(merged_df["Year"], merged_df["Total_Accidents"], color='red', marker='s', linestyle='--', label='Total Accidents')
ax2.set_ylabel("Total Accidents", color='red')
ax2.tick_params(axis='y', labelcolor='red')


# Title and Legend
plt.title("Comparison of Total Vehicles and Accidents by Year (Common Years Only)")
lines_1, labels_1 = ax1.get_legend_handles_labels()
lines_2, labels_2 = ax2.get_legend_handles_labels()
ax1.legend(lines_1 + lines_2, labels_1 + labels_2, loc="upper left")


plt.tight_layout()
plt.show()
```

The above graph shows that there is none linear-by-linear association between traffic flow and accident count directly.

**What increases the accident rates?** <br>

```python
fig, axes = plt.subplots(1, 4, figsize=(18, 5))

sns.countplot(x="Road_Surface_Conditions", data=meta_data, ax=axes[0],hue='Road_Surface_Conditions')
axes[0].set_title("Road Surface Conditions")

sns.countplot(x="Light_Conditions", data=meta_data, ax=axes[1],hue ="Light_Conditions")
axes[1].set_title("Light Conditions")

sns.countplot(x="Weather_Conditions", data=meta_data, ax=axes[2],hue='Weather_Conditions')
axes[2].set_title("Weather Conditions")

sns.countplot(x="Pedestrian_Crossing-Physical_Facilities", data=meta_data, ax=axes[3],hue='Pedestrian_Crossing-Physical_Facilities')
axes[3].set_title("Pedestrian Crossing / Physical_Facilities")

for ax in axes:
    ax.tick_params(axis='x', labelsize=9)
    for label in ax.get_xticklabels():
        label.set_rotation(45)
        label.set_horizontalalignment('right')

plt.tight_layout()
fig.subplots_adjust(bottom=0.3)
plt.show()
```

Accidents happened mostly in these environmental conditions:

- Road Surface: Dry
- Light Condition: Daylight, Street light present
- Weather condition: Fine without high winds
- Pedestrian Crossing / Physical Facilities: No physical crossing within 50 meters

Most of the accidents happened where the environment was in good conditions. So It also shows that environmental conditions might not be major impact on accidents. Other conditions should be also searched.

**Can we predict accident rates over time?**

We can forecast the accident count for the following years based on the past data. There is a forecast below for the next three years.

```python

# Yearly Total Accident Count
yearly_accidents =meta_data.groupby("Year").size().reset_index(name="Accident_Count")


print(yearly_accidents.head())
```

```python
# MODEL ARIMA

from statsmodels.tsa.arima.model import ARIMA

model = ARIMA(yearly_accidents["Accident_Count"], order=(1, 1, 1))
model_fit = model.fit()

# 3-year forecast
forecast = model_fit.forecast(steps=3)

# The last year & forecast years
last_year = yearly_accidents["Year"].max()
forecast_years = [last_year + i for i in range(1, 4)]

# forecast is a pandas Series, update index.
forecast_df = forecast.to_frame(name="Accident_Count")
forecast_df["Year"] = forecast_years
forecast_df.reset_index(drop=True, inplace=True)

combined_df = pd.concat([yearly_accidents, forecast_df], ignore_index=True)

plt.figure(figsize=(10, 6))
sns.lineplot(data=combined_df, x="Year", y="Accident_Count", label="Actualization & Prediction", marker="o")
plt.axvline(x=last_year, color='red', linestyle='--', label='Forecast Start')
plt.title("Estimated Annual Accidents (Actual and Estimated Data)")
plt.xlabel("Year")
plt.ylabel("Accidents")
plt.legend()
plt.grid(True)
plt.show()
```
